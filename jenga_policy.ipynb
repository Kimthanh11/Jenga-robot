### **⚙️ Create Robot Configuration**

```python
%%writefile /content/mjlab/src/mjlab/asset_zoo/robots/jenga/jenga_constants.py
from pathlib import Path
import mujoco
import copy
import xml.etree.ElementTree as ET
import mujoco

from mjlab import MJLAB_SRC_PATH
from mjlab.entity import Entity, EntityCfg, EntityArticulationInfoCfg
from mjlab.actuator import XmlActuatorCfg

JENGA_XML: Path = (
  MJLAB_SRC_PATH / "asset_zoo" / "robots" / "jenga" / "xmls" / "jenga.xml"
)
assert JENGA_XML.exists(), f"XML not found: {JENGA_XML}"

def get_spec() -> mujoco.MjSpec:
  return mujoco.MjSpec.from_file(str(JENGA_XML))

def get_spec_for_body(body_name: str, include_actuator: bool = False) -> mujoco.MjSpec:
  root = ET.parse(JENGA_XML).getroot()

  new_root = ET.Element("mujoco", {"model": body_name})

  for tag in ("compiler", "option", "default"):
    elem = root.find(tag)
    if elem is not None:
      new_root.append(copy.deepcopy(elem))

  worldbody = ET.SubElement(new_root, "worldbody")

  body = next(
    b for b in root.find("worldbody").findall("body")
    if b.get("name") == body_name
  )
  worldbody.append(copy.deepcopy(body))

  if include_actuator:
    actuator = root.find("actuator")
    if actuator is not None:
      new_root.append(copy.deepcopy(actuator))

  return mujoco.MjSpec.from_string(
    ET.tostring(new_root, encoding="unicode")
  )

def get_spec_for_floor():
    spec = mujoco.MjSpec()
    spec.worldbody.add_geom(
        name="floor",
        type=mujoco.mjtGeom.mjGEOM_PLANE,
        size=[0.3, 0.3, 0.01],
        rgba=[0.96, 0.93, 0.73, 1.0],
        friction=[0.8, 0.001, 0.0001],
    )
    return spec

def get_jenga_robot_cfg() -> dict[str, EntityCfg]:
    actuators = (
        XmlActuatorCfg(
            target_names_expr=("hook_slide",),
        ),
    )
    articulation = EntityArticulationInfoCfg(actuators=actuators)

    # Base dictionary containing the active pusher tool
    configs = {
        "floor": EntityCfg(
            spec_fn=get_spec_for_floor,
        ),
        "hook_slide": EntityCfg(
            spec_fn=lambda: get_spec_for_body("hook_slide", include_actuator=True),
            articulation=articulation,
        )
    }

    # Clean function factory to prevent Python lambda cell closure bugs inside nested loops
    def make_spec_fn(b_name):
        return lambda: get_spec_for_body(b_name)

    # Dynamically inject each block into the config mapping
    for layer in range(1, 10):
        for idx in range(1, 4):
            block_name = f"b{layer}_{idx}"
            configs[block_name] = EntityCfg(spec_fn=make_spec_fn(block_name))

    return configs

# if __name__ == "__main__":
#   import mujoco.viewer as viewer
#   robot = Entity(get_jenga_robot_cfg())
#   viewer.launch(robot.spec.compile())
```

### **📝 Create Environment Configuration**

This file contains the MDP (Markov Decision Process) components:
1. **Scene Config**: 64 parallel environments
2. **Actions**: Hook slide motor control with [-1, 1] command range
3. **Observations**: Hook position/velocity + selected block poses/velocities + tower stability features
4. **Rewards**: Successful block displacement + tower stability reward − action effort penalty
5. **Events**: Reset all blocks to original stacked positions
6. **Terminations**: Tower collapse

```python
%%writefile /content/mjlab/src/mjlab/asset_zoo/robots/jenga/env_cfg.py
"""Jenga task environment configuration."""

import torch
import mujoco
from mjlab.envs import ManagerBasedRlEnvCfg
from mjlab.managers.observation_manager import ObservationGroupCfg, ObservationTermCfg
from mjlab.managers.reward_manager import RewardTermCfg
from mjlab.managers.termination_manager import TerminationTermCfg
from mjlab.managers.event_manager import EventTermCfg
from mjlab.managers.scene_entity_config import SceneEntityCfg
from mjlab.scene import SceneCfg
from mjlab.sim import MujocoCfg, SimulationCfg
from mjlab.viewer import ViewerConfig
from mjlab.asset_zoo.robots.jenga.jenga_constants import get_jenga_robot_cfg
from mjlab.envs import mdp
from mjlab.envs.mdp.actions import JointEffortActionCfg


TARGET_BLOCK = "b7_3"
HOOK = "hook_slide"

TARGET_X_GOAL = 0.076    # 50% extraction
MAX_PUSH_FORCE = 1

# Grid layout physical dimensions
BLOCK_W = 0.05
BLOCK_L = 0.152
BLOCK_H = 0.03
SIDE_SPACING = 0.0505
START_Z = 0.0155
LAYER_HEIGHT = 0.0305
DROP_HEIGHT = 0.05

ALL_BLOCKS = [f"b{layer}_{idx}" for layer in range(1, 10) for idx in range(1, 4)]
_INITIAL_TARGET_X = None

def block_center(layer, position):
    z = START_Z + (layer - 1) * LAYER_HEIGHT
    internal_check_layer = layer + 1

    if internal_check_layer % 2 == 1:
        x = [-SIDE_SPACING, 0, SIDE_SPACING][position - 1]
        y = 0.0
        yaw = 0.0
    else:
        x = 0.0
        y = [SIDE_SPACING, 0, -SIDE_SPACING][position - 1]
        yaw = 1.57079632679

    return x, y, z, yaw


# Calculate the analytical fixed starting position once
X_START, Y_START, Z_START, YAW_START = block_center(7, 3) # for "b7_3"



def _hook_qpos(env):
    return env.sim.data.qpos[:, 0]

def _hook_qvel(env):
    return env.sim.data.qvel[:, 0]

def _hook_pos_w(env):
    pos = env.scene[HOOK].data.geom_pos_w
    if pos.dim() == 3:
        pos = pos[:, 0, :]
    return pos


def _hook_in_target_block_space(env):
    hook = env.scene[HOOK].data.geom_pos_w.squeeze(1)
    target = env.scene[TARGET_BLOCK].data.geom_pos_w.squeeze(1)

    rel = hook - target

    return rel

def _tower_center_of_mass(env):
    com = torch.zeros(env.num_envs, 3, device=env.device)
    count = 0
    for name in ALL_BLOCKS:
        if name == TARGET_BLOCK:
            continue
        com += env.scene[name].data.geom_pos_w.squeeze(1)
        count += 1
    return com / count




# ==============================================================================
# Scene Configuration
# ==============================================================================
def jenga_env_cfg(play: bool = False) -> ManagerBasedRlEnvCfg:
    # get_jenga_robot_cfg() already contains keys for "hook_slide", "b1_1", etc.
    scene_entities = get_jenga_robot_cfg()

    scene_cfg = SceneCfg(
        num_envs=16 if play else 8,
        extent=1.0,
        entities=scene_entities,
    )

    viewer_cfg = ViewerConfig(
        # origin_type=ViewerConfig.OriginType.ASSET_BODY,
        origin_type=ViewerConfig.OriginType.WORLD,
        entity_name=HOOK,
        body_name=HOOK,
        distance=2.0,
        elevation=20.0,
        azimuth=90.0,
    )

    sim_cfg = SimulationCfg(
        nconmax=512,
        njmax=512,
        mujoco=MujocoCfg(
            timestep=0.002,
            iterations=200,
        ),
    )

  # ==============================================================================
  # Actions
  # ==============================================================================
    actions = {
        "push_force": JointEffortActionCfg(
            entity_name=HOOK,
            actuator_names=("hook_slide",),
            scale=MAX_PUSH_FORCE,
        ),
    }

  # ==============================================================================
  # Observations
  # ==============================================================================
    actor_terms = {
        "hook_pos": ObservationTermCfg(
            func=lambda env: _hook_qpos(env).unsqueeze(-1),
        ),
        "hook_vel": ObservationTermCfg(
            func=lambda env: _hook_qvel(env).unsqueeze(-1),
        ),
        "target_xyz": ObservationTermCfg(
            func=lambda env: env.scene[TARGET_BLOCK].data.geom_pos_w.squeeze(1)
        ),
        "tower_com": ObservationTermCfg(
            func=lambda env: _tower_center_of_mass(env),
        ),
    }

    observations = {
        "actor": ObservationGroupCfg(
            terms=actor_terms,
            concatenate_terms=True,
            enable_corruption=not play,
        ),
        "critic": ObservationGroupCfg(
            terms=actor_terms,
            concatenate_terms=True,
            enable_corruption=False,
        ),
    }

  # ==============================================================================
  # Rewards
  # ==============================================================================
    def target_progress(env):
      current_x = env.scene[TARGET_BLOCK].data.geom_pos_w.squeeze(1)[:, 0]
      disp = (current_x - X_START).abs()
      progress = torch.clamp(disp / TARGET_X_GOAL, 0.0, 1.0)

      # print("disp max:", disp.max().item(), "progress max:", progress.max().item())

      return progress

    def target_success(env):
      current_x = env.scene[TARGET_BLOCK].data.geom_pos_w.squeeze(1)[:, 0]
      disp = (current_x - X_START).abs()
      success = disp >= TARGET_X_GOAL

      # if success.any():
      #     print(
      #         "success_envs =",
      #         success.sum().item(),
      #         "/",
      #         env.num_envs,
      #         "| disp max =",
      #         disp.max().item(),
      #         "| goal =",
      #         TARGET_X_GOAL,
      #     )

      return success & (~tower_collapsed(env))

    def target_success_reward(env):
        return target_success(env).float()

    def tower_collapsed(env):
        com = _tower_center_of_mass(env)
        com_too_far = torch.norm(com[:, :2], dim=-1) > 0.08
        # com_too_low = com[:, 2] < 0.01
        # return com_too_far | com_too_low
        return com_too_far



    def tower_com_penalty(env):
        com = _tower_center_of_mass(env)
        xy_shift = torch.norm(com[:, :2], dim=-1)
        return -xy_shift



    def hook_outside_target_face_penalty(env):
        hook = _hook_pos_w(env)
        target = env.scene[TARGET_BLOCK].data.geom_pos_w.squeeze(1)

        dy = torch.abs(hook[:, 1] - target[:, 1])
        dz = torch.abs(hook[:, 2] - target[:, 2])

        # vượt khỏi mặt block
        outside_y = torch.relu(dy - BLOCK_W / 2)
        outside_z = torch.relu(dz - BLOCK_H / 2)

        return -(outside_y + outside_z)

    rewards = {
        "target_progress": RewardTermCfg(func=target_progress, weight=20.0),
        "target_success": RewardTermCfg(func=target_success_reward, weight=50.0),
        "tower_com_stable": RewardTermCfg(func=tower_com_penalty, weight=0.5),
        "collapse": RewardTermCfg(
            func=lambda env: -tower_collapsed(env).float(),
            weight=5.0,
        ),
        "hook_outside_face": RewardTermCfg(
            func=hook_outside_target_face_penalty,
            weight=10.0,
        ),
    }

  # ==============================================================================
  # Events
  # ==============================================================================

    events = {
        "reset_hook_joint": EventTermCfg(
            func=mdp.reset_joints_by_offset,
            mode="reset",
            params={
                "asset_cfg": SceneEntityCfg(HOOK),
                "position_range": (0.0, 0.0),
                "velocity_range": (0.0, 0.0),
            },
        ),
    }

    for layer in range(1, 10):
        for idx in range(1, 4):
            name = f"b{layer}_{idx}"
            x, y, z, yaw = block_center(layer, idx)

            events[f"reset_{name}"] = EventTermCfg(
                func=mdp.reset_root_state_uniform,
                mode="reset",
                params={
                    "asset_cfg": SceneEntityCfg(name),
                    "pose_range": {
                        "x": (x, x),
                        "y": (y, y),
                        "z": (z + DROP_HEIGHT, z + DROP_HEIGHT),
                        "roll": (0.0, 0.0),
                        "pitch": (0.0, 0.0),
                        "yaw": (yaw, yaw),
                    },
                    "velocity_range": {
                        "x": (0.0, 0.0),
                        "y": (0.0, 0.0),
                        "z": (0.0, 0.0),
                        "roll": (0.0, 0.0),
                        "pitch": (0.0, 0.0),
                        "yaw": (0.0, 0.0),
                    },
                },
            )

  # ==============================================================================
  # Terminations
  # ==============================================================================
    terminations = {
        "target_success": TerminationTermCfg(
            func=target_success,
            time_out=False,
        ),
        "tower_collapsed": TerminationTermCfg(
            func=tower_collapsed,
            time_out=False,
        ),
    }

  # ==============================================================================
  # Environment Configuration
  # ==============================================================================
    return ManagerBasedRlEnvCfg(
        scene=scene_cfg,
        observations=observations,
        actions=actions,
        rewards=rewards,
        events=events,
        terminations=terminations,
        sim=sim_cfg,
        viewer=viewer_cfg,
        decimation=10,
        episode_length_s=30
    )
```

### **⚙️ Create RL Configuration**

This file defines the PPO (Proximal Policy Optimization) training parameters.

```python
%%writefile /content/mjlab/src/mjlab/asset_zoo/robots/jenga/rl_cfg.py
"""RL configuration for Jenga task."""

from mjlab.rl.config import (
  RslRlOnPolicyRunnerCfg,
  RslRlModelCfg,
  RslRlPpoAlgorithmCfg,
)


def jenga_ppo_runner_cfg() -> RslRlOnPolicyRunnerCfg:
  """Create RL runner configuration for Jenga task."""
  return RslRlOnPolicyRunnerCfg(
    actor=RslRlModelCfg(
      hidden_dims=(256, 128, 64),
      activation="elu",
      obs_normalization=True,
      distribution_cfg={
        "class_name": "GaussianDistribution",
        "init_std": 1.0,
        "std_type": "scalar",
      },
    ),
    critic=RslRlModelCfg(
      hidden_dims=(256, 128, 64),
      activation="elu",
      obs_normalization=True,
    ),
    algorithm=RslRlPpoAlgorithmCfg(
      value_loss_coef=1.0,
      use_clipped_value_loss=True,
      clip_param=0.2,
      entropy_coef=0.01,
      num_learning_epochs=5,
      num_mini_batches=4,
      learning_rate=1.0e-3,
      schedule="adaptive",
      gamma=0.99,
      lam=0.95,
      desired_kl=0.01,
      max_grad_norm=1.0,
    ),
    experiment_name="jenga",
    save_interval=50,
    num_steps_per_env=24,
    max_iterations=1000,
  )
```

### **📁 Locate Training Checkpoints**

After training, checkpoints are saved locally.

```python
import os
import re
from pathlib import Path

# Find the most recent training run
log_dir = Path("/content/mjlab/logs/rsl_rl/jenga")
if log_dir.exists():
  runs = sorted(log_dir.glob("*"), key=os.path.getmtime, reverse=True)
  if runs:
    latest_run = runs[0]
    print(f"✓ Latest training run: {latest_run.name}\n")

    # List checkpoints - sorted by iteration number
    checkpoints = list(latest_run.glob("model_*.pt"))
    if checkpoints:
      # Extract iteration number and sort numerically
      def get_iteration(ckpt):
        match = re.search(r"model_(\d+)\.pt", ckpt.name)
        return int(match.group(1)) if match else 0

      checkpoints = sorted(checkpoints, key=get_iteration)

      print(f"Found {len(checkpoints)} checkpoints:")
      for ckpt in checkpoints[-5:]:  # Show last 5
        size_mb = ckpt.stat().st_size / (1024 * 1024)
        iteration = get_iteration(ckpt)
        print(f"  • {ckpt.name} (iteration {iteration}, {size_mb:.2f} MB)")

      # Store the last checkpoint path
      last_checkpoint = str(checkpoints[-1])
      print(f"\n💾 Last checkpoint: {last_checkpoint}")
    else:
      print("⚠ No checkpoints found yet")
  else:
    print("⚠ No training runs found")
else:
  print("⚠ Log directory not found. Have you run training yet?")
```
